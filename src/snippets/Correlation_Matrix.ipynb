```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
```

```python
#load the data
market_df = pd.read_csv('../../data/y_train_HNMbC27.csv')
```

```python
#drop the values with na
new_market_df = market_df.dropna()
```

```python
#drop the index column and assign to new dataframe
new_market_df = new_market_df.iloc[:, 1:]
```

```python
#correlation matrix table with correlation values for all the 13 markets
new_market_df.corr().round(4)
```

```python
#correlation heatmap for visual analysis
sns.heatmap(new_market_df.corr().round(2))
plot.show()
```

From the above heatmap and matrix we can say that there is a correlation between some of the markets from 1st day to 1st week and 2nd week after the speech is delivered

```python
# retrieving the correlation matrix where the correlation is less than -0.7 and greater than 0.7
high_corr_matrix = new_market_df.corr()
high_corr_matrix = high_corr_matrix.unstack()
high_corr_matrix = high_corr_matrix[abs(high_corr_matrix) >= 0.7]

high_corr_matrix
```
